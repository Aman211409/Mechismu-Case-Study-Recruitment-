```python
info = {'Year':[2021,2022,2023,2024,2025,2026,2027,2028,2029,2030], 'YoY Growth Rate(%)':[6.5,6.7,6.9,6.9,7.9,7.5,7.2,7.32,6.82,6.38]}
electra = pd.DataFrame(info)
```

```python
electra
```

```python
plt.figure(figsize=(8,4))
plt.plot(electra['Year'], electra['YoY Growth Rate(%)'], linewidth=3, color="red", marker='o')
plt.xlabel("Year")
plt.ylabel('YoY Growth Rate(%)')
plt.style.use('classic')
plt.xlim(2021, 2030)
plt.ylim(6,10)
plt.title("Electra's YoY Growth", pad=10)
for x,y in zip(electra['Year'],electra['YoY Growth Rate(%)']):

    label = "{:.2f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center', fontsize=9) 
```
